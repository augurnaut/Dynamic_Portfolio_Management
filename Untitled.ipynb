```python
raw_cmc_stocks_3 = pd.read_csv(
    Path('./HistoricalData_1629327952562.csv'), 
    infer_datetime_format=True, 
    parse_dates=True)

raw_cmc_stocks_3.head()
```

```python
raw_cmc_stocks_4 = pd.read_csv(
    Path('./CMC200.csv'), 
    infer_datetime_format=True, 
    parse_dates=True)

raw_cmc_stocks_4.head()
```

```python
AAPL_stock = raw_cmc_stocks_1[['Date','Close/Last']].copy()
MSFT_stock = raw_cmc_stocks_2[['Date','Close/Last']].copy()
CSCO_stock = raw_cmc_stocks_3[['Date','Close/Last']].copy()
CMC200_index = raw_cmc_stocks_4[['Date','Close']].copy()


AAPL = pd.DataFrame(AAPL_stock[['Date','Close/Last']])
MSFT = pd.DataFrame(MSFT_stock[['Date','Close/Last']])
CSCO = pd.DataFrame(CSCO_stock[['Date','Close/Last']])
CMC200 = pd.DataFrame(CMC200_index[['Date','Close']])

print(AAPL.head())
print(MSFT.head())
print(CSCO.head())
print(CMC200.head())
```

```python
#data_pull = [AAPL, MSFT, CSCO, CMC200]

raw_data = pd.merge([AAPL, MSFT, CSCO, CMC200], join='Date', axis=1)
raw_data.columns = ["AAPL", "MSFT", "CSCO","CMC200"]
raw_data.head()

#raw_data = pd.merge(AAPL, MSFT, CSCO, on="Date")
```

```python
raw_data["AAPL"] = raw_data["AAPL"].str.replace('$', '').astype(float)
raw_data["MSFT"] = raw_data["MSFT"].str.replace('$', '').astype(float)
raw_data["CSCO"] = raw_data["CSCO"].str.replace('$', '').astype(float)
```

```python
type(raw_data["CMC200"][1])
```

```python
raw_data['AAPL'] = 3 * raw_data['AAPL']
print(raw_data['AAPL'])
```

```python
raw_data['MSFT'] = 4 * raw_data['MSFT']
raw_data['CSCO'] = 5 * raw_data['CSCO']
```

```python
raw_data.head()
```

```python
raw_data['portfolio_market_cap'] = raw_data['AAPL'] + raw_data['MSFT'] + raw_data['CSCO']
```
