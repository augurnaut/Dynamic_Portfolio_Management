```python
raw_cmc_stocks_2 = pd.read_csv(
    Path('./HistoricalData_1629327921152.csv'), 
    infer_datetime_format=True, 
    parse_dates=True)

raw_cmc_stocks_2.head()
```

```python
raw_cmc_stocks_3 = pd.read_csv(
    Path('./HistoricalData_1629327952562.csv'), 
    infer_datetime_format=True, 
    parse_dates=True)

raw_cmc_stocks_3.head()
```

```python
raw_cmc_stocks_44 = pd.read_csv(
    Path('./CMC200.csv'), 
    infer_datetime_format=True, 
    parse_dates=True)

raw_cmc_stocks_44 = raw_cmc_stocks_44.sort_values(by=['Date'], ascending = False)

raw_cmc_stocks_44.reset_index(drop=True, inplace=True)

raw_cmc_stocks_4 = raw_cmc_stocks_44[['Date','Close']]

raw_cmc_stocks_4.head(50)
```

```python
AAPL_stock = raw_cmc_stocks_1[['Date','Close/Last']].copy()
MSFT_stock = raw_cmc_stocks_2[['Date','Close/Last']].copy()
CSCO_stock = raw_cmc_stocks_3[['Date','Close/Last']].copy()
CMC200_index = raw_cmc_stocks_4[['Date','Close']].copy()

AAPL = pd.DataFrame(AAPL_stock[['Date','Close/Last']])
MSFT = pd.DataFrame(MSFT_stock[['Date','Close/Last']])
CSCO = pd.DataFrame(CSCO_stock[['Date','Close/Last']])
CMC200 = pd.DataFrame(CMC200_index[['Date','Close']])

AAPL.columns = ['Date','Close']
MSFT.columns = ['Date','Close']
CSCO.columns = ['Date','Close']
CMC200.columns = ['Date','Close']

AAPL['Date']= pd.to_datetime(AAPL['Date'])
AAPL['Date'] = AAPL['Date'].dt.strftime("%Y-%m-%d")

CSCO['Date']= pd.to_datetime(CSCO['Date'])
CSCO['Date'] = CSCO['Date'].dt.strftime("%Y-%m-%d")

MSFT['Date']= pd.to_datetime(MSFT['Date'])
MSFT['Date'] = MSFT['Date'].dt.strftime("%Y-%m-%d")

CMC200['Date']= pd.to_datetime(CMC200['Date'])
CMC200['Date'] = CMC200['Date'].dt.strftime("%Y-%m-%d")

print(AAPL)

#MSFT = pd.DataFrame(MSFT_stock[['Date','Close/Last']])
#CSCO = pd.DataFrame(CSCO_stock[['Date','Close/Last']])
#CMC200 = pd.DataFrame(CMC200_index[['Date','Close']])

#print(AAPL.head())
#print(MSFT.head())
#print(CSCO.head())
#print(CMC200.head())
```

```python
data_pull = [AAPL, MSFT, CSCO, CMC200]

raw_data = pd.concat(data_pull, axis = 1)
raw_data.columns = ["Date","AAPL", "Date_2","MSFT","Date_3", "CSCO","Date_4","CMC200"]
raw_data.tail()

#raw_data = pd.merge(AAPL, MSFT, CSCO, on="Date")
```

```python
raw_portfolio_data = raw_data[['Date','AAPL','MSFT','CSCO','CMC200']].copy()
raw_portfolio_data.tail()
```

```python
raw_portfolio_data["AAPL"] = raw_portfolio_data["AAPL"].str.replace('$', '').astype(float)
raw_portfolio_data["MSFT"] = raw_portfolio_data["MSFT"].str.replace('$', '').astype(float)
raw_portfolio_data["CSCO"] = raw_portfolio_data["CSCO"].str.replace('$', '').astype(float)
```

```python
print(type(raw_portfolio_data['CMC200'][2]))
print(type(raw_portfolio_data['AAPL'][2]))
raw_portfolio_data.tail()
```

```python
raw_portfolio_data['AAPL'] = 3 * raw_portfolio_data['AAPL']
print(raw_portfolio_data['AAPL'])
```

```python
raw_portfolio_data['MSFT'] = 4 * raw_portfolio_data['MSFT']
raw_portfolio_data['CSCO'] = 5 * raw_portfolio_data['CSCO']
```

```python
raw_portfolio_data.tail(30)
```

```python
raw_portfolio_data['portfolio_market_cap'] = raw_portfolio_data['AAPL'] + raw_portfolio_data['MSFT'] + raw_portfolio_data['CSCO']
```

```python
raw_portfolio_data = raw_portfolio_data.sort_values(by=['Date'], ascending = True)

raw_portfolio_data.reset_index(drop=True, inplace=True)

raw_portfolio_data.tail(50)
```

```python
raw_portfolio_data.describe()
```

```python
raw_portfolio_data.dropna(subset = ["CMC200"], inplace=True)
raw_portfolio_data.head(50)
```

```python
raw_portfolio_data.describe()
```

```python
raw_portfolio_data['CMC200 Cumulative Returns'] = ((raw_portfolio_data['CMC200'] - raw_portfolio_data['CMC200'][1])/raw_portfolio_data['CMC200'][1]) * 100
raw_portfolio_data['Portfolio Cumulative Returns'] = ((raw_portfolio_data['portfolio_market_cap'] - raw_portfolio_data['portfolio_market_cap'][1])/raw_portfolio_data['portfolio_market_cap'][1]) * 100

raw_portfolio_data['CMC200 Cumulative Returns']  = round(raw_portfolio_data['CMC200 Cumulative Returns'], 4)
raw_portfolio_data['Portfolio Cumulative Returns']  = round(raw_portfolio_data['Portfolio Cumulative Returns'], 4)

raw_portfolio_data.tail()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline 

plt.subplots(figsize=(9, 6))

X = raw_portfolio_data['CMC200 Cumulative Returns'] 
Y = raw_portfolio_data['Portfolio Cumulative Returns']
Z = raw_portfolio_data['Date'] 

plt.plot(X, label="CMC200 Cumulative Returns")
plt.plot(Y, label="Stock Portfolio Cumulative Returns")

plt.legend(loc='upper left')

plt.title("Portfolio Comparison to Crypto Index", fontsize=16, fontweight='bold')

plt.xlabel("Trading Day")
plt.ylabel("Cumulative Return")

#ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))

plt.savefig('Portfolio Comparison to Crypto Index.png')
plt.show()
```
