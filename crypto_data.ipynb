```python
import pandas as pd
from pathlib import Path

%matplotlib inline 
```

```python
# All data is from binance for the past 3 years

btc_data_df = pd.read_csv(
    Path('./Resources/Binance_BTCUSDT_d.csv'), 
    skiprows = 1, 
    infer_datetime_format=True, 
    parse_dates=True)

btc_data_df.head()
```

```python
btc_close_data = btc_data_df.drop(columns=['unix','open','high','low','Volume BTC'])

btc_close_data.head()
```

```python
btc_close_data.set_index('date', inplace=True)

btc_close_data.head()
```
