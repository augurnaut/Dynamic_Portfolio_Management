```python
crypto_close_data.columns = ['ADA','BNB','BTC','DOT','ETH','LINK','LTC','VET','XLM','XRP']

crypto_close_data.head()
```

```python
crypto_close_data.columns = ['ADA','BNB','BTC','DOT','ETH','LINK','LTC','VET','XLM','XRP']

crypto_close_data.head()
```

```python
crypto_daily_returns = crypto_close_data.pct_change().dropna()

crypto_daily_returns
```

```python
crypto_cumulative_returns = (1+crypto_daily_returns).cumprod()

crypto_cumulative_returns
```

```python
crypto_cumulative_returns.plot(title = 'Cumulative Returns', figsize = (50,10))
```
